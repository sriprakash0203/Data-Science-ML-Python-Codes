```python
def preprocess(x):
         x=re.sub('[^a-zA-Z ]', '', x)
         x=x.lower()
         x=x.split()  
         x=[word  for word in x if word not in set(stopwords.words('english'))]  
         x=[ps.stem(word) for word in x] 
         x=" ".join(x)
         return x
    
    
```
