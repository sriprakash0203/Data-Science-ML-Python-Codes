```python
import matplotlib.pyplot as plt    #importing plot libary
plt.style.use("seaborn-darkgrid")  #background dark and brid
%matplotlib inline                 #to save in notebook
```

```python
x = [1,2,3] #x-coridnates
y= [5,6,7]  #y-coridnates
```

```python
plt.figure(figsize=(10,5))    #to get layout space
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x,y)  #only points shown
plt.plot(x,y)     #draws lines
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x,y,color="red",s=50) #change points color and size
plt.plot(x,y,color="black",linestyle="--",linewidth=1.5)   #change color,line pattern,size
plt.xlabel('time',size=25)   #label x-axiz and size of label
plt.ylabel('money$',size=25,) #label y-axiz and size of label
plt.title('SRI',size=20,y=1.1) #label title,size of label,spacing b/w plot and title
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x,y,color="red",s=50)
plt.scatter(1,5,color="blue",s=50)  #to change only 1 point 
plt.plot(x,y,color="black",linestyle="--",linewidth=1.5)
plt.xlabel('time',size=25)
plt.ylabel('money$',size=25,)
plt.title('SRI',size=20,y=1.1)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x,y,color="red",s=50)
plt.scatter(1,5,color="blue",s=50)
plt.plot(x,y,color="black",linestyle="--",linewidth=1.5)
plt.xlabel('time',size=25)
plt.ylabel('money$',size=25,)
plt.title('SRI',size=20,y=1.1)
plt.text(2,6,"Sriprakash",size=20) #to give a name near the point
```

```python
plt.figure(figsize=(10,5),facecolor="red")   #faceclolor to change the backgroung color
plt.scatter(x,y,color="red",s=50)
plt.scatter(1,5,color="blue",s=50)
plt.plot(x,y,color="black",linestyle="--",linewidth=1.5)
plt.xlabel('time',size=25)
plt.ylabel('money$',size=25,)
plt.title('SRI',size=20,y=1.1)
plt.text(2,6,"Sriprakash",size=20)
```

```python
plt.figure(figsize=(10,5),facecolor="green")
plt.scatter(x,y,color="red",s=50)
plt.scatter(1,5,color="blue",s=50)
plt.plot(x,y,color="black",linestyle="--",linewidth=1.5)
plt.xlabel('time',size=25)
plt.ylabel('money$',size=25,)
plt.title('SRI',size=20,y=1.1)
plt.text(2,6,"Sriprakash",size=20)
plt.ylim(0,10) #change yaxis limit
plt.xlim(0,20) #change xaxiz limit
```

```python
plt.figure(figsize=(10,5),facecolor="green")
plt.scatter(x,y,color="red",s=50)
plt.scatter(1,5,color="blue",s=50)
plt.plot(x,y,color="black",linestyle="--",linewidth=1.5)
plt.xlabel('time',size=25)
plt.ylabel('money$',size=25,)
plt.title('SRI',size=20,y=1.1)
plt.text(2,6,"Sriprakash",size=20)
plt.ylim(-5,10)
plt.xlim(-5,20)
plt.hlines(0,-1,10)
plt.vlines(0,-2,10)
plt.savefig("pythonplot.jpg")  #to save plot in local
```

```python
plt.figure(figsize=(10,5),facecolor="red")
plt.scatter(x,y,color="red",s=50)
plt.scatter(1,5,color="blue",s=50)
plt.plot(x,y,color="black",linestyle="--",linewidth=1.5)
plt.xlabel('time',size=25)
plt.ylabel('money$',size=25,)
plt.suptitle('SRIPRAKASH L',size=25,y=1.2)  #Bigger title (super-title)
plt.title('SRI',size=20,y=1.1)
plt.text(2,6,"Sriprakash",size=20)
plt.ylim(-5,10)
plt.xlim(-5,20)
plt.hlines(0,-1,10)
plt.vlines(0,-2,10)
plt.savefig("pythonplot.jpg")
```

# Histogram

```python
plt.figure(figsize=(6,6),facecolor='white')
plt.hist(mtcars['hp'],facecolor="black",edgecolor='red',bins=8)
#creates histogram with 8 bins and colours filled init
plt.xlabel('hp')
```

```python
#BoxPlot
#help(plt.boxplot)
plt.figure(figsize=(6,6),facecolor='white')
plt.boxplot(mtcars['hp'],vert=False)  #if true thn vertical plot
```

# Seaborn


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
plt.figure(figsize=(8,6))
df['size'].value_counts().plot(kind='pie',autopct="%1.1f%%")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['total_bill'],df['tip'])
```

# Univariate-->"One variable" data visualization

```python
#strip plot

sns.stripplot(y='tip',data=df,jitter=True)   #scatter plot
plt.ylabel('tip')
plt.show()
```

```python
sns.stripplot(y='tip',data=df,jitter=False)   #straight plot
plt.ylabel('tip')
plt.show()
```

```python
df.columns
```

```python
#grouping with strip plot
plt.figure(figsize=(8,6))
sns.stripplot(x='size',y='tip',data=df) 
```

```python
#grouping with box plot
plt.figure(figsize=(8,6))
sns.boxplot(x='size',y='tip',data=df)    #box 5 in plot
```

```python
df[df['size']==5]['tip']
```

```python
df[df['size']==5]['tip'].quantile([0.25,0.50,0.75])
```

```python
sns.stripplot(x='time',y='tip',data=df)
```

```python
sns.boxplot(x='time',y='tip',data=df)
```

```python
sns.stripplot(x='day',y='tip',hue='sex',data=df)
```
